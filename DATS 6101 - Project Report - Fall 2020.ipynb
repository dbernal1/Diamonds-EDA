# 1) Introduction

Diamond is a solid form of the element carbon with its atoms are arranged in a crystal structure called diamond cubic. This carbon has been one of the most sough-after and expensive gemstones for various industiral and decorative purposes since the ancient time. The hardness, brilliance, and sparkle of diamonds make them unsurpassed as gems. 

The value of a diamond is determined by what are commonly known as the Four "C's" of a Diamond. These C's stand for carat weight, clarity, color, and the quality of its cut. Most diamonds' color ranges from clear to yellow to brown with the clear being the most expensive. Cut refers to how a diamondâ€™s facets interact with light. It is determined by symmetry, proportion and polish. Carat denotes the weight of a diamond, not the size. Diamond clarity is a measure of the purity and rarity of the stone, graded by the visibility of these characteristics under 10-power magnification. A stone is graded as flawless if, under 10-power magnification, no inclusions (internal flaws) and no blemishes (external imperfections) are visible.

**5.3 ANOVA**

For the ANOVA analysis we focused on checking the relationship between the 5 cut type groups and their prices. Before proceeding with the analysis, we implemented the 3 assumption checks: 

1) Independence:

* Tested during the design of the study. All groups must be mutually exclusive, i.e. in our data, a diamond can only belong in one of the cut type group. 

2) Normality: 

* We checked for normality using the **Shapiro-walk** test from scipy.stats.

3) Homogeniety of Variance

* we checked for equal variance among groups using the **Barlette's and Levene's** test



**5.3.1 Shapiro-Walk Normality test**

Ho : data is normally distributed         
Ha: data is not normally distributed

Out put was a **p-value: 0.917**. We accepted the null hypothesis that our data was normally distributed.

**5.3.2 Homogeneity of Variance - Barlett's Test**

Ho : Our 5 groups categorized by cut type have equal variance      
Ha: Our 5 groups categorized by cut type don not have equal variance    

our p-value was less than 0.05, which indicated that we reject our null hypothesis that our groups have equal variances. Since the Barlett's test was a bit sensitive to variance differnces, we also checked using Leven's test. However, both test confirmed that our groups do not have equal variances. This was mainly due to the number of samples in each of our 5 groups were hugely different, resulting a biased variances. Hence, parametric test such as ANOVA, t-test can not be perfomred in our samples. As a result, we chose to perform **Kruskal-Wallis** test for the non-parametric analysis.

**5.3.3 Kruskal-Wallis Non-Parametric Test**

Ho: The mean price of the 5 cut groups are the same     
Ha: The mean prices of the 5 cut groups are different



Kruskal-Wallis analysis, shows that the mean prices for the different cut groups are different.
